```
import csv
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
english_file = "/content/gdrive/My Drive/Colab Notebooks/swiggy/english_translation.txt"
telugu_file = "/content/gdrive/My Drive/Colab Notebooks/swiggy/telugu_indic_translation_no_pp.txt"
```

```
eng_tokenizer = Tokenizer()

data_eng = open(english_file).read()

corpus_eng = data_eng.lower().split("\n")

print(corpus_eng)

print(len(corpus_eng[:5000]))

eng_tokenizer.fit_on_texts(corpus_eng)

word_index = eng_tokenizer.word_index

print(word_index)

total_words_eng = len(eng_tokenizer.word_index) + 1

print(total_words_eng)

# create input sequences using list of tokens
input_sequences_eng = []

for line in corpus_eng[:5000]:
  #print(line)
  token_list = eng_tokenizer.texts_to_sequences([line])[0]
  #print(token_list)
  for i in range(1, len(token_list)):
    n_gram_sequence = token_list[:i+1]
    #print(n_gram_sequence)
  #print(n_gram_sequence)
  input_sequences_eng.append(n_gram_sequence)
    
print(len(input_sequences_eng))
```
